###**Task 2:**
Find the shape, rank and size of the tensors in 1 above
***

* **Vecor Tensor** also known as a "rank-1" tensor - it is similar to a list of values.  It has **one axis**
* **Scalar Tensor** also known as a "rank-0" tensor - contains a single value.  It has **no axes**
* **Matrix Tensor** also known as a "rank-2" tensor - contains a single value.  It has **two axes**

```python
import tensorflow as tf
import numpy as np
#creating a vector tensor - vector tensor a
vt = tf.constant([2,3,4,5])
st = tf.constant(5)
mt = tf.constant([[2,3,4,5],[5,6,7,8]])
print(f'VECTOR TENSOR: \n{vt}')
print(f'The shape and rank of the VECTOR: Shape:{vt.shape} | Rank:{vt.ndim} | Size:{tf.size(vt)}')
print(f'SCALAR TENSOR: \n{st}')
print(f'The shape and rank of the SCALAR: Shape:{st.shape} | Rank:{st.ndim} |Size:{tf.size(st)} ')
print(f'MATRIX TENSOR: \n{mt}')
print(f'The shape and rank of the MATRIX: Shape:{mt.shape} | Rank:{mt.ndim} Size:{tf.size(mt)} ')
```

###Other forms of creating Tensors
* Creating a random matrix tensor, use the set seed if you need to reproduce the same random values
* Creating a tensor with many dimensions

```python
#Creating multidimensional random tensor
mtDimensionalRandomTensor = tf.constant(tf.random.uniform(shape=[20]), shape=(1,1,1,1,20))
print(f'Multi Dimensional TENSOR: \n{mtDimensionalRandomTensor}')
```

```python
myRandomTensor = tf.random.Generator.from_seed(42)
myRandomTensor = myRandomTensor.normal(shape=(3,4))
print(f'RANDOMLY GENERATED MATRIX WITH SEED: \n{myRandomTensor}')
print(f'The shape of the Tensor above is: {myRandomTensor.shape}')
print(f'The above tensor has {myRandomTensor.ndim} dimensions')
```

```python
#Creating a tensor with many dimensions
fourDimensionalTensor =tf.ones(shape=(3,2,5,4))
print(f'FOUR DIMENSIONAL TENSOR: \n{fourDimensionalTensor}')
print(f'The shape of the Tensor above is: {fourDimensionalTensor.shape}')
print(f'The above tensor has {fourDimensionalTensor.ndim} dimensions')
```

###**Task 3:**
Create two tensors containing random values between 0 and 1 with the shape [5,300]

```python
randomTensor1 = tf.random.uniform(shape=[5,300], minval=0, maxval =1)
randomTensor2 = tf.random.uniform(shape=[5,300], minval=0, maxval =1)
randomTensor1, randomTensor2
```

###**Task 4:**
Multiply the two tensors created in task 3 using matrix multiplication

```python
#check the shapes of the two tensors
randomTensor1.shape, randomTensor2.shape
```

**We will need to reshape or transpose one of them to meet the rules of matrix multiplication**

```python
#reshape randomTensor1 and them perform matrix multiplication
reShaped = tf.reshape(randomTensor1, shape=(300,5))
myProduct = tf.matmul(reShaped,randomTensor2) 
print(f'The product of randomTensor1 of shape{reShaped.shape} * randomTensor2 of shape{randomTensor2.shape} is: \n{myProduct}')
```

###**Task 5:**
Multiply the two tensors created in task 3 using dot product

```python
tf.tensordot(randomTensor1,randomTensor2, axes=0)
```
